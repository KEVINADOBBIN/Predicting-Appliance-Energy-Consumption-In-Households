```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# 2. Load Dataset
df = pd.read_excel("energydata_complete.xlsx")  # or .csv if needed
print(df.head())
```

```python
# 3. Initial Data Check
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
# 4. Exploratory Data Analysis
px.histogram(df, x="Appliances", nbins=50, title="Distribution of Appliance Energy Consumption").show()
px.scatter(df, x="T_out", y="Appliances", trendline="ols", title="Outdoor Temperature vs Appliance Usage").show()
corr = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr, cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 5. Feature Engineering
X = df.drop(columns=["date", "Appliances", "rv1", "rv2"])
y = df["Appliances"]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 6. Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# 7. Model 1: Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
```

```python
# 8. Model 2: Random Forest Regressor
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

```python
# 9. Evaluation
def evaluate_model(y_true, y_pred, name):
    mse = mean_squared_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    print(f"{name} - MSE: {mse:.2f}, R: {r2:.2f}")

evaluate_model(y_test, y_pred_lr, "Linear Regression")
evaluate_model(y_test, y_pred_rf, "Random Forest")
```

```python
# 10. Feature Importance (Random Forest)
importances = rf.feature_importances_
features = df.drop(columns=["date", "Appliances", "rv1", "rv2"]).columns
importance_df = pd.DataFrame({"Feature": features, "Importance": importances}).sort_values(by="Importance", ascending=False)
px.bar(importance_df, x="Feature", y="Importance", title="Feature Importance (Random Forest)").show()
```
